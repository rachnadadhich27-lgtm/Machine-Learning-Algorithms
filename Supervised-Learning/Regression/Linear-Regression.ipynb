```python
import numpy as np
import matplotlib.pyplot as plt


# Set the seed for reproducibility
np.random.seed(42)

# Number of samples
num_samples = 100

# Generate random x values (features)
x_train = 2 * np.random.rand(num_samples, 1)

# Generate corresponding y values with a linear relationship (y = 4 + 3x + noise)
true_slope = 3
true_intercept = 4
noise = np.random.randn(num_samples, 1)

y_train = true_intercept + true_slope * x_train + noise

# Plotting the data
plt.figure(figsize = (8, 5))
plt.scatter(x_train, y_train, color = "blue", label = "Training data", alpha = 0.7)
plt.title("Dummy Linear Regression Data")
plt.xlabel("x_train")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape is:", x_train.shape)
print("Target shape is:", y_train.shape)
```

```python
#initial parameters
w = 0
b = 0

learning_rate = 0.1
trials = 1000

m = len(x_train)

cost_history = []

# GRADIENT DESCENT 
for trials in range(trials):

    total_cost = 0
    dw = 0
    db = 0
    for i in range(m):

        x = x_train[i][0]
        y = y_train[i][0]
        # PREDICTION
        y_pred = w * x + b
        # ERROR
        error = y_pred - y
        # COST CALCULATION
        total_cost = total_cost + (error ** 2)
        # GRADIENTS
        dw = dw + (error * x)
        db = db + error
    # FINAL COST
    cost = total_cost / (2 * m)
    # AVERAGE GRADIENTS
    dw = dw / m
    db = db / m
    # UPDATE PARAMETERS
    w = w - learning_rate * dw
    b = b - learning_rate * db
    # STORE COST
    cost_history.append(cost)
        
 # FINAL PARAMETERS
print("\nFinal Learned Parameters")

print("Weight (w) =", w)

print("Coefficient (b) =", b)
# PREDICTIONS
y_predictions = []
for i in range(m):
    x = x_train[i][0]

    y_pred = w * x + b

    y_predictions.append(y_pred)


#  list to numpy array
y_predictions = np.array(y_predictions)

# PLOT TRAINING DATA
plt.figure(figsize=(8,5))
plt.scatter(x_train,
            y_train,
            color="blue",
            label="Training Data")

# PLOT REGRESSION LINE
plt.plot(x_train,
         y_predictions,
         color="red",
         linewidth=2,
         label="Regression Line")


plt.xlabel("x_train")
plt.ylabel("y_train")

plt.title("Linear Regression using Gradient Descent")

plt.legend()

plt.grid(True)

plt.show()
# PLOT COST FUNCTION

plt.figure(figsize=(8,5))

plt.plot(range(trials), cost_history)

plt.xlabel("trials")
plt.ylabel("Cost")

plt.title("Cost Function")

plt.grid(True)

plt.show()
```
