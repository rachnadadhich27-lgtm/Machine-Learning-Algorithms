```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Number of samples and features
num_samples = 100
num_features = 3

# Generate random x values (features)
x_train = 2 * np.random.rand(num_samples, num_features)

# Define true weights (slopes) and intercept
true_weights = np.array([[3], [2], [1]])  # shape: (3, 1)
true_intercept = 4

# Generate noise
noise = np.random.randn(num_samples, 1)

# Compute y values
y_train = true_intercept + x_train @ true_weights + noise  # @ is matrix multiplication

# Plotting using only first feature for visualization
plt.figure(figsize = (8, 5))
plt.scatter(x_train[:, 0], y_train, color = "green", label = "Training data (vs 1st feature)", alpha = 0.7)
plt.title("Dummy Linear Regression Data (3 features)")
plt.xlabel("x_train[:, 0] (1st feature)")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape is:", x_train.shape)
print("Target shape is:", y_train.shape)
```

```python
# Add bias column
X = np.c_[np.ones((num_samples,1)), x_train]
# Total parameters
n = X.shape[1]
#Initializing parameters randomly
w = np.random.randn(n,1)
# Learning rate
learning_rate = 0.01
# Number of trials
trials = 1000
# Store cost values
cost_history = []
# Gradient Descent Loop
for i in range(trials):
    # Prediction
    y_pred = X @ w
    # Error
    error = y_pred - y_train
    # Cost Function
    cost = (1/(2 * num_samples)) * np.sum(error ** 2)
    # Store cost
    cost_history.append(cost)

    # Gradient
    gradients = (1/(num_samples))* (X.T @ error)
    # Update parameters
    w= w - learning_rate* gradients
    

print("\nFinal Parameters")
print("\nBias (b):")
print(w[0][0])

print("\nWeights:")
print(w[1:])

# Final prediction
y_final = X @ w

# PREDICTED GRAPH
plt.figure(figsize=(8,5))
plt.scatter(x_train[:,0],
            y_final,
            color="red",
            alpha=0.5,
            label="Predicted Data")

plt.xlabel("First Feature")
plt.ylabel("Target")
plt.title("Predicted value")
plt.legend()
plt.grid(True)
plt.show()
# COST FUNCTION GRAPH

plt.figure(figsize=(8,5))

plt.plot(range(iterations), cost_history)

plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function vs Iterations")
plt.grid(True)

plt.show()
```
