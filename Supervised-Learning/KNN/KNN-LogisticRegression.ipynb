```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification


# Set seed
np.random.seed(42)

# Generate a classification dataset
x, y = make_classification(
    n_samples=200,
    n_features=2,          # Only 2 useful features for visualization
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    class_sep=0.8,         # Low separation makes it harder
    flip_y=0.1,            # Add noise (10% label flipping)
    random_state=42
)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x[y == 0][:, 0], x[y == 0][:, 1], color="red", label="Class 0", alpha=0.6)
plt.scatter(x[y == 1][:, 0], x[y == 1][:, 1], color="blue", label="Class 1", alpha=0.6)
plt.title("Dummy Binary Classification Data (for Logistic Regression)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape:", x.shape)
print("Target shape:", y.shape)
```

```python
k=5
y_pred=[]

for i in range(x.shape[0]):
    difference=x[i]-x
    distance=np.sqrt(np.sum((difference**2),axis=1))
    index=np.argsort(distance)[1:k+1]

    nearest_neigh=y[index]
    values, counts = np.unique(nearest_neigh, return_counts=True)
    most_frequent = values[np.argmax(counts)]

    y_pred.append(most_frequent)
    
accuracy=np.mean(y_pred==y)*100
y_pred=np.array(y_pred)
```

```python
# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x[y_pred== 0][:, 0], x[y_pred== 0][:, 1], color="red", label="Class 0", alpha=0.6)
plt.scatter(x[y_pred == 1][:, 0], x[y_pred == 1][:, 1], color="blue", label="Class 1", alpha=0.6)
plt.title("Dummy Binary Classification Data (for Logistic Regression)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape:", x.shape)
print("Target shape:", y_pred.shape)
```
