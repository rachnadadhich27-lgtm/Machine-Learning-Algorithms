```python
import numpy as np
import matplotlib.pyplot as plt


# Set the seed for reproducibility
np.random.seed(42)

# Number of samples
num_samples = 100

# Generate random x values (features)
x_train = 2 * np.random.rand(num_samples, 1)

# Generate corresponding y values with a linear relationship (y = 4 + 3x + noise)
true_slope = 3
true_intercept = 4
noise = np.random.randn(num_samples, 1)

y_train = true_intercept + true_slope * x_train + noise

# Plotting the data
plt.figure(figsize = (8, 5))
plt.scatter(x_train, y_train, color = "blue", label = "Training data", alpha = 0.7)
plt.title("Dummy Linear Regression Data")
plt.xlabel("x_train")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape is:", x_train.shape)
print("Target shape is:", y_train.shape)
```

```python
k=5
y_pred=[]

for i in range(x_train.shape[0]):
    difference=x_train[i]-x_train
    distance=np.sqrt(np.sum((difference**2),axis=1))
    index=np.argsort(distance)[1:k+1]

    nearest_neigh=y_train[index]
    
    y_pred.append(np.mean(nearest_neigh))
    
y_pred=np.array(y_pred)
```

```python
# Plotting the data
plt.figure(figsize = (8, 5))
plt.scatter(x_train, y_train, color = "blue", label = "Training data", alpha = 0.7)
plt.scatter(x_train, y_pred, color = "red", label = "Training data", alpha = 0.7)
plt.title("Dummy Linear Regression Data")
plt.xlabel("x_train")
plt.ylabel("y_pred")
plt.legend()
plt.grid(True)
plt.show()

```
